### Transformation

The `Fits` class offers a range of methods for transforming digital images. It strives to preserve the World Coordinate System (WCS) information, if present, throughout these transformations.

```python
from myraflib import Fits

fits = Fits.sample()
fits.show()
```

### shift

The `Fits` class allows you to shift a digital image by a specified amount. This functionality is useful for aligning images where rotation is not a concern.

```python
shifted_fits = fits.shift(100, 100)
shifted_fits.show()
```

### Rotate

The `Fits` class provides a method to rotate the digital image by a specified angle, given in radians.

```python
from math import radians

rotated_fits = fits.rotate(radians(45))
rotated_fits.show()
```

### Crop

The `Fits` class includes a method for cropping the image. This method returns the specified subwindow of the image, defined by the lowest x and y coordinates, width, and height. Parameters required are:

- **x-coordinate** of the lowest corner
- **y-coordinate** of the lowest corner
- **width** of the subwindow
- **height** of the subwindow


```python
cropped_fits = fits.crop(100, 100, 500, 500)
cropped_fits.show()
```

### Bin

The `Fits` class features a `bin` method for binning the digital image. Binning can be useful for various photometric techniques, where combining adjacent pixels into larger bins helps in data analysis and noise reduction.

```python
binned_fits = fits.bin(10)
binned_fits.show()
```

### Align

The `Fits` class includes a method for aligning two FITS images. This functionality helps in registering images to ensure that corresponding features are accurately matched, which is crucial for comparative analysis and combining data.

```python
to_be_aligned = fits.shift(150, 120).rotate(radians(45))
to_be_aligned.show()
```

```python
aligned_fits = to_be_aligned.align(fits)
aligned_fits.show()
```
