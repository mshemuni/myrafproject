# show

The `Fits` object also provides functionality to visualize the data it contains.

```python
from myraflib import Fits

fits = Fits.sample()
fits.show()
```

The `show` method offers two optional parameters: `scale` and `sources`. The `scale` parameter is a boolean, and when set to `True`, the data will be z-scaled before visualization. The `sources` parameter can be either a `pandas.DataFrame` or an `astropy.table.Table` (but not a `QTable`), containing `xcentroid` and `ycentroid` fields to overlay on the image.

```python
fits.show(scale=False)
```

```python
import pandas as pd

sources = pd.DataFrame(
    {
        'xcentroid': list(range(200, 800, 100)),
        'ycentroid': list(range(200, 800, 100)),
    }
    
)

fits.show(sources=sources)
```
