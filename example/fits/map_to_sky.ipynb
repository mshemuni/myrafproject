# map_to_sky
`Fits` can retrieve sources available in the region of the FITS data from Simbad

```python
from myraflib import Fits
from matplotlib import pyplot as plt
from astropy.visualization import ZScaleInterval
from random import choices

fits = Fits.sample()
fits
```

```python
sky_map = fits.map_to_sky()
sky_map
```

```python
plt.imshow(ZScaleInterval()(fits.data()), cmap="gray_r")

for each in choices(sky_map[["name", "xcentroid", "ycentroid"]].values.tolist(), k=10):
    plt.text(each[1], each[2], each[0], color='red', fontsize=8)
    plt.scatter(each[1], each[2], color="red", s=10)

plt.show()
```
